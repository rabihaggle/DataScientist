```python
import pandas as pd
import numpy as np
df_p = pd.read_csv('p_data.csv', delimiter = ';', engine = 'python')
df_w = pd.read_csv('w_data.csv', delimiter = ';', engine = 'python')
```

```python
df_p.head()
```

```python
df_w.head()
```

```python
df_inner = pd.merge(df_p, df_w, how = 'inner', on = 'ID' , indicator= 'union')
df_inner.head()
```

```python
df_inner[df_inner == 'both'].count()
```

```python
df_inner.Target.unique()
```

```python
df_inner.Target.replace([' <=50K',' <=50K.'], '<=50K', inplace = True)
df_inner.Target.replace([' >50K',' >50K.'], '>50K', inplace = True)
```

```python
df_inner.Target.unique()
```

```python
df_group_Target = df_inner.Target.groupby(df_inner.Target).count().reset_index(name = 'count')
df_group_Target['Porcentaje'] = (df_group_Target['count']/df_group_Target['count'].sum())*100

df_group_Target.head()
```

```python
df_group_Sex = df_inner[['Target','Sex']]
df_group_Sex = df_group_Sex.groupby(['Sex','Target']).size().reset_index(name= 'count')
df_group_Sex['Porcentaje'] = (df_group_Sex['count'] / df_group_Sex['count'].sum())*100
optiones = ['>50K']
df_group_Sex = df_group_Sex[df_group_Sex['Target'].isin(optiones)]
df_group_Sex.head()
```

```python
df_group_Sex = df_inner[['Target','Education']]
df_group_Sex = df_group_Sex.groupby(['Education','Target']).size().reset_index(name= 'count')
df_group_Sex['Porcentaje'] = (df_group_Sex['count'] / df_group_Sex['count'].sum())*100
optiones = ['>50K']
df_group_Sex = df_group_Sex[df_group_Sex['Target'].isin(optiones)]
df_group_Sex.head()
```

```python
df_group_Sex = df_inner[['Target','Race']]
df_group_Sex = df_group_Sex.groupby(['Race','Target']).size().reset_index(name= 'count')
df_group_Sex['Porcentaje'] = (df_group_Sex['count'] / df_group_Sex['count'].sum())*100
optiones = ['>50K']
df_group_Sex = df_group_Sex[df_group_Sex['Target'].isin(optiones)]
df_group_Sex.head()
```
